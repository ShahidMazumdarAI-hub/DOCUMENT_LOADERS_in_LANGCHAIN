```python
#making a text file for executing TextLoader wala kaam
with open('cricket2.txt', 'w') as f:
   f.write('''Cricket is a bat-and-ball game that is played between two teams of eleven players on a field, at the centre of which is a 22-yard (20-metre; 66-foot) pitch with a wicket at each end, each comprising two bails (small sticks) balanced on three stumps. Two players from the batting team, the striker and nonstriker, stand in front of either wicket holding bats, while one player from the fielding team, the bowler, bowls the ball toward the striker's wicket from the opposite end of the pitch. The striker's goal is to hit the bowled ball with the bat and then switch places with the nonstriker, with the batting team scoring one run for each of these swaps. Runs are also scored when the ball reaches the boundary of the field or when the ball is bowled illegally.''')
```

```python
# DOC Loaders is one of the components of RAG used for making RAG based applications
# DOC Loaders ke help se aap kisi v source se DOCument load kar paate ho
# there are many types of DOC LOADER in L.CHAIN :-> TEXTLOADER, PyPDFLoader, WebBaseLoader, CSVLoader etc etc
# TEXTLOADER:-> text files ko uthaakar usko document ke form me banaakar LANGCHAIN me leke aataa hai

from langchain_community.document_loaders import TextLoader #Saare ke saare Document loaders in L.CHAIN aapko langchain_community me hi milenge

#sabse pehle apko ek loader obj banaana padta hai which is a document loader object and is obj ko initialize karte tym aap parameters bhej sakte ho like aapke txt file ka path
loader = TextLoader('cricket2.txt')

#ab is doc loader obj ke paas ek func hota hai jiska naam hai LOAD, which will load ur text file as a DOCUMENT in ur memory, i will store this in a variable(docs)
docs = loader.load()
print(docs) #we will see a LIST in the o/p
```

```python
print(docs[0].page_content)
```

```python
#now i can use this docs document from the above in various ways
#i can form a chain ki mai is text(docs me jo text hai) ko extract kar raha hu and is text ko mai bhej raha hu LLM ke paas for generating the summary of this text

from langchain_community.document_loaders import TextLoader #Saare ke saare Document loaders in L.CHAIN aapko langchain_community me hi milenge
from langchain_openai import ChatOpenAI
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate

HUGGINGFACEHUB_ACCESS_TOKEN = " " #for security and integrity purposes, i didn't share my API key. The entire program can also be done using OpenAI platform by using OpenAI API key

llm = HuggingFaceEndpoint(
    repo_id="HuggingFaceH4/zephyr-7b-beta",
    task="text-generation",
    huggingfacehub_api_token=HUGGINGFACEHUB_ACCESS_TOKEN

)

model = ChatHuggingFace(llm=llm)

prompt = PromptTemplate(
    template='Write a summary for the following topic- \n {topic}',
    input_variables=['topic']
)

parser = StrOutputParser()






loader = TextLoader('cricket2.txt')


docs = loader.load()
print(docs)

chain = prompt | model | parser

print(chain.invoke({'topic':docs}))
```

```python
print(chain.invoke({'topic':docs[0].page_content}))
```

***For SCANNED IMAGES WALA PDFs ETC ETC THERE ARE OTHER DOC LOADERS IN L.CHAIN Like UNSTRUCTUREDPDF LOADER etc etc****

***if U are having a folder jiske andar MULTIPLE PDFs hai or multiple text files hai, aur aapko in sabhi ko sath me load karna hai L.CHAIN me, that time we use a CLASS named as DIRECTORY_LOADER***

```python
# we are having 2 pdfs(ML topic) and game text file in our SAMPLE Folder of our G.COLAB, We will try loading them together from the SAMPLE FOLDER
from langchain_community.document_loaders import PyPDFLoader, DirectoryLoader

loader = DirectoryLoader(         #here loader is the object of DirectoryLoader class
    path='sample_data', #aapka directory ka path, means folder name jaha par aapne multiple PDFs ya text files ko store karke rakha hai
    glob='*.pdf', #is sample data folder ke andar se mujhe kaun si file load karna hai, for this we have a parameter(glob) where we can explain the pattern ki sirf pdf files humehumare sample data folder se extract karke do ya phir only csv files hume humare sample folder se extract karke do
    loader_cls=PyPDFLoader  #aapko yeh batana hoga ki apka loader kaun se class ka hai
)

docs = loader.load()
print(len(docs))

```

***(3). WEB BASE LOADER***

```python
# is DOC LOADER ke help se aap kisi v web page ka content utha karke L.CHAIN me load kar sakte ho aur phir uske upar Qns kar sakte ho
# WEB BASE LOADER uses 2 Python libraries:-> REQUEST(with this aap HTTP request karte ho to that web page), BEAUTIFULSOUP(jiske help se aap us Web page ka HTML structure samajhte ho aur usko textual format me leke aate ho Langchain me)
from langchain_community.document_loaders import WebBaseLoader

url = "https://in.puma.com/in/en/pd/deviate-nitro-4-mens-max-cushion-road-running-shoes/312123?swatch=01"
loader = WebBaseLoader(url)

docs = loader.load()
print(docs)
```

```python
print(docs[0].page_content)
```

```python
#now we can ask Qns on the web page loaded above, lets see the code for this below:
from langchain_community.document_loaders import WebBaseLoader
from langchain_openai import ChatOpenAI
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate

llm = HuggingFaceEndpoint(
    repo_id="HuggingFaceH4/zephyr-7b-beta",
    task="text-generation",
    huggingfacehub_api_token=HUGGINGFACEHUB_ACCESS_TOKEN

)

model = ChatHuggingFace(llm=llm)

prompt = PromptTemplate(
    template='Answer the following Question \n {Question} from the following - \n {text}',
    input_variables=['Question','text']
)

parser = StrOutputParser()

url = "https://in.puma.com/in/en/pd/deviate-nitro-4-mens-max-cushion-road-running-shoes/312123?swatch=01"
loader = WebBaseLoader(url)

docs = loader.load()

chain = prompt | model | parser
print(chain.invoke({'Question':'What is the colour of this shoe?', 'text':docs[0].page_content}))
```

```python
print(chain.invoke({'Question':'can i get a good resale value for this shoe?', 'text':docs[0].page_content}))
```
